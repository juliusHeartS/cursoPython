```python
# Respuesta al ejercicio de preguntas
pregunta1 = "¿Cuál es su nombre?"
pregunta2 = "¿En qué ciudad nació?"
pregunta3 = "¿Cuántos años tiene?"

nombre = input(pregunta1)
ciudad = input(pregunta2)
edad = input(pregunta3)
print("Hola ", nombre, "! de ", ciudad, ". Espero que a sus ", edad, " años se sienta muy bien!" )
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Leyendo la imagen
img = mpimg.imread('mario.png')

# Mostrando la imagen
plt.imshow(img)

# Mostrando la gráfica
plt.show()
```

```python

# Mostrando la imagen en escala de grises
from PIL import Image
img = Image.open('mario.png')
gray_img = img.convert('L')

plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
import numpy as np

# Create a random grayscale image as a 2D numpy array
img = np.random.rand(100, 100)

# Display the image as grayscale
plt.imshow(img, cmap='gray')

# Show the plot
plt.show()
```
